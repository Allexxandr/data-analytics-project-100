```python
import pandas as pd
import requests
import json
```

```python
df_visits = pd.read_csv('https://drive.google.com/uc?export=download&id=1QosQQ4RRNR9rkL4t7sB707h2Uy0XfYJe')
```

```python
df_registrations = pd.read_csv('https://drive.google.com/uc?export=download&id=1AeQz0kaSgz0lxYSDtuNm36muhy5fRCzZ')
```

```python
response = requests.get('https://data-charts-api.hexlet.app/visits?begin=2023-03-01&end=2023-09-01')
```

```python
print(json.dumps(response.json()[:10], indent=4))
```

```python
df_visits = df_visits[~df_visits['user_agent'].str.contains('bot', case=False, na=False)]
```

```python
df_visits['date'] = pd.to_datetime(df_visits['date'])
df_visits['date'] = df_visits['date'].dt.date
```

```python
visits_grouped = df_visits.groupby(['date', 'platform']).size().reset_index(name='visits')
```

```python
visits_grouped.head(30)
```

```python
df_registrations['date'] = pd.to_datetime(df_registrations['date'])
df_registrations['date'] = df_registrations['date'].dt.date
registrations_grouped = df_registrations.groupby(['date', 'platform']).size().reset_index(name='registrations')
```

```python
registrations_grouped.head(50)
```

```python
combined_data = pd.merge(visits_grouped, registrations_grouped, on=['date', 'platform'], how='outer').fillna(0)
```

```python
combined_data['conversion'] = (combined_data['registrations'] / combined_data['visits']).round(4)
```

```python
combined_data.columns = ['date_group', 'platform', 'visits', 'registrations', 'conversion']
```

```python
combined_data.head(50)
```

```python
ads_df = pd.read_csv('https://drive.google.com/uc?export=download&id=12vCtGhJlcK_CBcs8ES3BfEPbk6OJ45Qj')
```

```python
ads_df.head(50)
```

```python
ads_df['date'] = pd.to_datetime(ads_df['date'])
ads_df['date'] = ads_df['date'].dt.date
```

```python
ads_grouped = ads_df.groupby(['date', 'utm_campaign'])['cost'].sum().reset_index()
```

```python
ads_grouped.head(100)
```

```python
merged_data = pd.merge(combined_data, ads_grouped, left_on=['date_group'], right_on=['date'], how='left')
```

```python
merged_data.head(50)
```

```python
merged_data['cost'] = merged_data['cost'].fillna(0)
merged_data['utm_campaign'] = merged_data['utm_campaign'].fillna('none')
```

```python
merged_data = merged_data.rename(columns={'date_group': 'date','cost': 'cost','utm_campaign': 'utm_campaign' })
```

```python
merged_data = merged_data[['date', 'visits', 'registrations', 'cost', 'utm_campaign']]
```
